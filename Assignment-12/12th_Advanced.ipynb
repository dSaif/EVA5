## Create a config file

```
%%writefile cifar.yml
name: TinyImageNet
save_dir: saved/
seed: 1
target_device: 0

arch:
    type: TinyImageNetS12Model
    args: {}

augmentation:
    type: TinyImageNetAlbumentations
    args: {}

data_loader:
    type: TinyImageNetDataLoader
    args:
        batch_size: 256
        data_dir: data/
        nworkers: 6
        shuffle: True

lr_finder:
    epochs: 20

criterion: cross_entropy_loss

lr_scheduler:
    max_lr_at_epoch: 12
    type: OneCycleLR
    args:
        max_lr: 0.1

optimizer:
    type: SGD
    args:
        lr: 0.001
        momentum: 0.95
        weight_decay: 0.0005

training:
    epochs: 50
```

## Import sodium

```
from sodium.utils import load_config
import sodium.runner as runner
```

## Create a runner with the config and train the model

```
!nvidia-smi
```

```
config = load_config('cifar.yml')
runner = runner.Runner(config)
runner.setup_train(tsai_mode=True)
```

# Model Summary

```
runner.print_summary(input_size=(3, 64, 64))
```

# Model Visualization

```
runner.print_visualization(input_size=(3, 64, 64))
```

# Find the Optimal LR

```
runner.find_lr()
```

```
runner.train(use_bestlr=True)
```

## Plot the metrics

```
runner.plot_metrics()
```
