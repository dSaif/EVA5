## Create a config file

```
%%writefile cifar.yml
name: CIFAR10_MyNet
save_dir: saved/
seed: 1
target_device: 0

arch:
    type: CIFAR10S8Model
    args: {}

augmentation:
    type: CIFAR10Albumentations
    args: {}

data_loader:
    type: CIFAR10DataLoader
    args:
        batch_size: 128
        data_dir: data/
        nworkers: 4
        shuffle: True

criterion: cross_entropy_loss

lr_scheduler:
    type: OneCycleLR
    args:
        max_lr: 0.1

optimizer:
    type: SGD
    args:
        lr: 0.001
        momentum: 0.95
        weight_decay: 0.0005

training:
    epochs: 20
```

## Import sodium

```
from sodium.utils import load_config
import sodium.runner as runner
```

## Create a runner with the config and train the model

```
config = load_config('cifar.yml')
runner = runner.Runner(config)
runner.train(tsai_mode=True)
```

## Plot the metrics

```
runner.plot_metrics()
```

## Plot Grad-CAM

```
target_layers = ["layer1", "layer2", "layer3", "layer4"]
runner.plot_gradcam(target_layers=target_layers)
```
