## Create a config file

```
%%writefile quiz.yml
name: QuizModel
save_dir: saved/
seed: 1
target_device: 0

arch:
    type: QuizDNNModel
    args: {}

augmentation:
    type: CIFAR10Albumentations
    args: {}

data_loader:
    type: CIFAR10DataLoader
    args:
        batch_size: 512
        data_dir: data/
        nworkers: 4
        shuffle: True

criterion: cross_entropy_loss

lr_scheduler:
    type: OneCycleLR
    args:
        max_lr: 0.5

optimizer:
    type: SGD
    args:
        lr: 0.001
        momentum: 0.95
        weight_decay: 0.0005

training:
    epochs: 20
```

## Import sodium

```
from sodium.utils import load_config
import sodium.runner as runner
```

## Create a runner with the config and train the model

```
config = load_config('quiz.yml')
runner = runner.Runner(config)
runner.train(tsai_mode=True)
```

## Plot the metrics

```
runner.plot()
```
